```stata
*Include a model for each of your hypotheses
*If it's a nominal variable that isn't binary, you have to put a coefficient for each category that's not going to be omitted. Example, race if excluding white you have to put coefficient betas for all other races

est clear 

*Model 1
quietly eststo: reg inccont slfemp 

*Model 2
quietly eststo: reg inccont slfemp educ i.race
```

```stata
esttab, stats(r2 N, labels("R-Squared" "N")) cells(b(star fmt(3)) se(fmt(3) par)) /// 
  nobase mlabels ("Model 1" "Model 2") starlevels(* .05 ** .01 *** .001) ///
  coeflabels (educ "education" 2.race "black" 3.race "other race") noconstant
```
