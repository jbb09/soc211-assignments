## Walkthroughs on Non-Linear Models

Let's say I want to investigate the relationship between loan repayment (y_lo_inc_rpy) and college net price for low-income students (netprice4lowincome).

### Creating spline example (1):

Splines are basically wriggly/curvy lines with knots aka points where there is a shift in magnitude/direction. Mathematically, they are represented by piecewise polynomial functions.

Cubic splines are represented mathematically by piecewise polynomial functions:
$ y = \begin{cases}
{2x^3 + x^2 + 4x - 1} & x\leq 0 \\
{2x^3 - 10} & 0\leq x\leq 100 \\
{9x^3 - 7} & 100\leq x
\end{cases}
$

Charlie showed everyone in class how to create a linear spline with one knot. Linear splines are linear fitted lines with knots aka points where there are shifts in direction or magnitude.

A linear spline is represented mathematically by piecewise linear functions:
$ y = \begin{cases} 
      {3x-1} & x\leq 0 \\
      {100-33x} & 0\leq x\leq 100 \\
      {9x-10} & 100\leq x 
   \end{cases}
$



Let's say I anticipate a linear spline relationship between loan repayment and college net price (both of which are continuous variables). Without background knowledge or intuition to guide me, I decide to use quantiles to represent the knot. Keep in mind- a spline model requires that both the y-variable and x-variable are continuous.

```stata
*Percent loan repayment is the dependent variable, whereas college net price is the independent variable

sum y_lo_inc_rpy netprice4lowincome
```

```stata
xtile netquant = netprice4lowincome, nq(5)

*I am creating a new variable here (called netquant) that will consist of quantiles for college net price

*I specified that I want 5 quantiles for this variable with the option "nq(5)"

*5 quantile means that there would be these categories:
    * 1 = 0-20% percentile
    * 2 = 20-40% percentile
    * 3 = 40-60% percentile
    * 4 = 60-80% percentile
    * 5 = 80-100% percentile
```

#### How do you read percentiles?:
If you fall into the 10% percentile for instance, that means your value/score is higher than 10% of the sample. If you fall into the 99% percentile, your value/score is higher than 99% of the sample.

```stata
sum netprice4lowincome, detail
tab netquant
```

```stata
%head y_lo_inc_rpy netprice4lowincome netquant
```

```stata
*Let's say that I believe there is one knot, and that occurs in the middle quantile (3rd quantile)
    *I would first get the average/mean for college net prices that fall into the 3rd quantile (40-60th percentile)

sum netprice4lowincome if netquant==3
```

```stata
*Now let's create the linear spline variables

mkspline netprice1 8773.032 netprice2 = netprice4lowincome

histogram netprice1, name(hist1)
histogram netprice2, name(hist2)
```

```stata
*Now let's run the regression output for this spline model and examine the graph for it

reg y_lo_inc_rpy netprice1 netprice2
predict yhat

*Graph the fited spline lines for college net price
tw (scatter y_lo_inc_rpy netprice4lowincome) (line yhat netprice4lowincome, sort color(red) lwidth(thick))
```

#### Concluding interpretation:

The result looks terrible! The data definitely is not well-represented by this spline.

### Creating spline example (2):

```stata
regress y_lo_inc_rpy netprice3*
```

#### Concluding interpretation:
A spline, either linear or cubic, is inappropriate for modeling the relationship between percent loan repayment and college net price for low-income students. We have a high R-squared for the cubic spline model, but this simply represents an example of a model that is overfitting the data.

What is overfitting?
* According to IBM, "[i]n machine learning, overfitting occurs when an algorithm fits too closely or even exactly to its training data, resulting in a model that can’t make accurate predictions or conclusions from any data other than the training data" (https://www.ibm.com/topics/overfitting).

For your homework assignment, a result where there is overfitting occurring like this is fine. You would simply state how the spline model is not a good representation of your predictions/hypothesis.

### Creating quadratic model example:

Now we are using quadratic formulas to model the relationship between loan repayment (y_lo_inc_rpy) and college net price for low-income students (netprice4lowincome). Keep in mind- once again, this is appropriate when both your y-variable and x-variable of interest are both continuous.

Here is an example of a quadratic function:
$ y = 5x^2 + 3x + 10 $

```stata
gen netpricequad = netprice4lowincome^2
```

```stata
regress y_lo_inc_rpy netprice4lowincome netpricequad
```

```stata
tw (scatter y_lo_inc_rpy netprice4lowincome, sort) ///
(lfit y_lo_inc_rpy netprice4lowincome, color(red) lwidth(thick)) ///
(qfit y_lo_inc_rpy netprice4lowincome, color(green) lwidth(thick)), ///
legend(off) name(quadratic, replace)
```

#### Concluding interpretation:

There is barely a U-curve for the quadratic model (notice the terrible p-value for the quadratic coefficient in the regression table and the crappy magnitude).

However, the best way to write up a formal interpretation of what is going on here is to first discuss the linear fitted line. Then, plug in hypothetical numbers into your equation, and describe what is going on. You may want to use more than one hypothetical number.

For instance, the equation for this model would be as followed:

---

$\hat{Y} = \alpha + \beta_{n1} \cdot x_{n1} + \beta_{n2} \cdot {x_{n2}^2} $

$\hat{Y}$ is the predicted/estimated value of percent loan repayment

$\alpha$ is the constant (otherwise known as the intercept term)

$x_{n1}$ is college net price for low-income students

${x_{n2}^2}$ is college net price for low-income students squared (quadratic term)

---

Let's say that I predict that there is a declining rate of percent loan repayment when the net college price for low-income students hit $15,000. I'm going to plug that into this model.

First, according to Stata, this is the estimated regression model:
$\hat{Y} = 80.79764 + (-.0020901) \cdot x_{n1} + (4.44e-08) \cdot {x_{n2}^2}$

Now, I will plug in $15,000 for college net price:
1. Y = 80.79764 + (-.0020901) * 15000 + (4.44e-08) * (15000)^2
2. Y = 59% loan repayment

Formal interpretation of this: When net college price for low-income students reaches $15,000, there is around 59% loan repayment occurring. After this, there is a decline in the negative rate at which net college price affects percent loan repayment.
* NOTE: I know there is a decline occurring based on eyeballing the model at the point of my hypothetical number.

### Creating log-dependent model example:

This is a natural log function:
$x = ln(y)$

This is log function:
$x = log(y)$

In Stata, we are working with the following:
$y = e^x$
OR
$y = 10^x$

For the log-dependent model, your y- and x- variables should again be continuous variables (also known as interval-ratio variables).

```stata
*Note: Since this is a log-DEPENDENT model, we are transforming the DEPENDENT variable in this.

gen loggedrepay=log10(y_lo_inc_rpy)
```

```stata
reg loggedrepay netprice4lowincome
```

As Charlie stated during lecture, graphs are the best way for you to interpret what is going on, and it helps your reader understand what is going on as well.

```stata
predict yhat3
replace yhat3=10^(yhat3)

*Creating fitted line for regular linear model
quietly reg y_lo_inc_rpy netprice4lowincome
predict yhat4, xb
```

```stata
tw (scatter y_lo_inc_rpy netprice4lowincome) ///
(line yhat4 netprice4lowincome, sort color(red) lwidth(thick)) ///
(line yhat3 netprice4lowincome, sort color(blue) lwidth(thick)), legend(off)
```

#### Concluding interpretation:

The red line (log-dependent model) is almost the same as the blue line (linear model). Just by eyeballing this, I would argue that the linear model is a better fit for the data based on parsimony. If the log-dependent model had captured more of what is occurring in the data, I would be able state that the log-dependent model is a better presentation of the sample.

Let's pretend that the log-dependent model is amazing. I eyeballed it, and I think that it's a great fit for my data. How would I then interpret my log-dependent model? Simple- you would just exponentiate the estimated coefficients by constant e or 10.
* As a result of exponentiating the estimated coefficient, you receive the multiplicative factor for every one-unit increase in the independent variable.

Since I did log of 10 rather than natural log in this model, I would exponentiate the coefficient of college net price by 10.
* -7.83e-06 becomes 10^(-7.83e-06), which is equal to approximately 0.99.

Formal interpretation: For every one unit increase in college net price for low-income students, percent loan repayment decreases by a factor of 0.99, or 1%.
* Note: Multiplying a number by 0.99 is the same as decreasing it by 1% (100-99 = 1). On the flip side, multiplying a number by 1.01 is the same as increasing it by 1%.

### Turning continuous independent variable into sequential dummies:

Let's say that I want to turn college net price into 6 brackets based on vibes. I'm hypothesizing that there are maybe differences in average percent loan repayment at 3k, 5k, 7k, 9k, 11k, 13k, and 15k college net prices, so I want to create brackets based on this.

```stata
sum netprice4lowincome, detail
```

```stata
egen netprice_vibes = cut(netprice4lowincome), at(3000,5000,7000,9000,11000,13000,15000)
```

```stata
tab netprice_vibes

*The cut function here looks at the left-hand of the interval (so from 0-3000, 3000-5000, etc). Be mindful of this.
```

```stata
reg y_lo_inc_rpy i.netprice_vibes
```

```stata
*You will create a margins plot for easier interpretation of what you are seeing.

qui margins i.netprice_vibes, atmeans
marginsplot, ytitle(estimated percent loan repayment)
```

#### Concluding interpretation:

I would formally interpret this by focusing on the means of each bracket. For instance, at the 0-3k bracket for college net price, the average loan repayment is at 70%. Moving from this to the 3-5k bracket, we see a decline in average loan repayment to 65%.

In terms of comparing this sequential dummy model to the linear model, compare the R-square for both the linear model vs the sequential dummy model, but also keep in mind theory. At the end of the day, theory dominates.
